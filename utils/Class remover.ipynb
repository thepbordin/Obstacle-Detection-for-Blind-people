## Import Json


```python
import json
```

```python
filepath = "cocoFormat.json"
removeclassid = [2, 3, 4]
```

## Load COCOFormat Json file

```python
obj = json.load(open(filepath))
anno = obj["annotations"]
cag = obj["categories"]
img = obj["images"]
```

## Find Annotation ids in innotations to remove

```python
#Remove from annotation

aremove = []
for n in range(len(anno)):
    if anno[n]["category_id"] in removeclassid:
        aremove.append(n)
aremove.sort(reverse=True) # fix index reordering
print(f"Annotation ids to remove: {aremove[:10]}")
```

## Find Category ID to remove in Category list

```python
cremove = []
for n in range(len(cag)):
    if cag[n]["id"] in removeclassid:
        cremove.append(n)
cremove.sort(reverse=True) # fix index reordering
print(f"Anno ids to remove: {cremove}")
```

## Remove from targets

```python
for ar in aremove:
    del anno[ar]
    print(f"Removed ID: {ar+1}")
for cr in cremove:
    del cag[cr]
    print(f"Removed CAG ID: {cr}")
```

```python
len(anno)
```

## Save file (Overwrite to the target)

```python
obj["annotations"] = anno 
obj["categories"] = cag
```

```python
with open(filepath, 'w') as f:
    json.dump(obj, f)
```

#### Retrieve all image ids from annotations

```python
imgid = [(lambda x: anno[x]["image_id"])(x) for x in range(len(anno))]
```

#### Remove Duplicated ids

```python
imgid = list(set(imgid))
imgid.sort()
```

#### Retrieve image list 

```python
imglist = [(lambda x: img[x]["id"])(x) for x in range(len(img))]
```

#### Find unlabeled images

```python
unlabel = set(imglist) - set(imgid)
unlabel = list(unlabel)
unlabel[-10:]
```

```python
imgremove = []
for n in range(len(img)):
    if img[n]['id'] in unlabel:
        imgremove.append(n)
imgremove.sort(reverse=True)
imgremove[-10:]
```

```python
for r in imgremove:
    del img[r]
for row in range(len(img)):
    assert img[row]["id"] not in unlabel
```
