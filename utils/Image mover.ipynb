```python
import glob
import os.path as op
import shutil
import os
```

# Extract images into train/valid/test folders

```python
images = glob.glob("imgs/*.*")
len(images)
```

```python
trainAnnos = glob.glob("train/images/*.txt")
validAnnos = glob.glob("valid/images/*.txt")
testAnnos = glob.glob("test/images/*.txt")
len(trainAnnos)+len(validAnnos)+len(testAnnos)
```

```python
validAnnos
```

```python

for source, new_path in [[trainAnnos, "train/images/"], [validAnnos, "valid/images/"], [testAnnos, "test/images/"]]:
    image_path = []
    # Get image Path from annotation file name
    for annopath in source: 
        file_name = op.splitext(op.basename(annopath))[0]
        image_name = glob.glob(f"imgs/{file_name}.*")
        if image_name != []:
            image_path.append(image_name[0]) 
            # print(image_name[0])
    for old_path in image_path:
        shutil.copy(old_path, op.join(new_path, op.basename(old_path)))
        # os.rename(old_path, op.join(new_path, op.basename(old_path)))
        # print(old_path, new_path)
```
